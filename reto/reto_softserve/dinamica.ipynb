```python
df=pd.read_html(str(morelostable))
# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head())
```

```python
with open('altitud.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f, delimiter='\t')
    headers = next(reader)
    headers = [header.strip().lower().replace(' ', '_') for header in headers]
    Row = namedtuple('Row', headers)
    morelos_altitud = []
    for row in reader:
        row = Row(*row)
        morelos_altitud.append(tuple([row.municipio, row.altitud_msnm]))
        
altitud = pd.DataFrame.from_records(morelos_altitud, columns=['Municipio', 'Altitud'])
altitud['Altitud'] = altitud['Altitud'].str.replace(',', '')
altitud['Altitud'] = altitud['Altitud'].apply(pd.to_numeric)
```

```python
poblacion = pd.read_csv('morelospoblacion.csv')
poblacion = poblacion.loc[:,['Municipio\n​','Población\n(2020)​']]

import re

# define the regex pattern
pattern = r'\d+'

# define a function to apply the regex pattern to each element of the column
def extract_numbers(string):
    numbers = re.findall(pattern, string)
    return numbers[0] if numbers else None

# apply the function to the second column of the dataframe
poblacion['Población\n(2020)​'] = poblacion['Población\n(2020)​'].apply(extract_numbers)

poblacion.columns = ['Municipio', 'Poblacion']
poblacion['Poblacion'] = poblacion['Poblacion'].astype(int)
```

```python
altitud.sort_values(by='Municipio', ascending=True, inplace=True)
poblacion.sort_values(by='Municipio', ascending=True, inplace=True)
```

```python
# clean the Municipio column in altitud dataframe
altitud['Municipio'] = altitud['Municipio'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
altitud['Municipio'] = altitud['Municipio'].str.lower().str.strip()

# clean the Municipio column in poblacion dataframe222
poblacion['Municipio'] = poblacion['Municipio'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
poblacion['Municipio'] = poblacion['Municipio'].str.lower().str.strip()

# merge the dataframes
merged_df = altitud.merge(poblacion, on='Municipio', how='inner')
```
